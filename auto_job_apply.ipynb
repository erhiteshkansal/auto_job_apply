```python
import fitz  # PyMuPDF

def extract_text_from_pdf(pdf_path):
    # Open the PDF file
    with fitz.open(pdf_path) as pdf:
        # Extract text from each page
        text = ""
        for page_num in range(pdf.page_count):
            page = pdf[page_num]
            text += page.get_text()
    return text

# Load your PDF and extract text
pdf_path = "resume.pdf"
resume_text = extract_text_from_pdf(pdf_path)
```
